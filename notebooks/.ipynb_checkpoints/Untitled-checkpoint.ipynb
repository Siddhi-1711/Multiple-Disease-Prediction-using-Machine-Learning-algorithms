```python
import pandas as pd
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Accuracy scores for SVM, Logistic Regression, and Random Forest
svm_accuracy = 0.7912087912087912
lr_accuracy = 0.8852459016393442
rf_accuracy = 0.8852459016393442

# Create a DataFrame for the accuracies
accuracies_df = pd.DataFrame({
    'Algorithm': ['SVM', 'Logistic Regression', 'Random Forest'],
    'Accuracy': [svm_accuracy, lr_accuracy, rf_accuracy]
})

# Plot the comparison
plt.figure(figsize=(8, 6))
plt.bar(accuracies_df['Algorithm'], accuracies_df['Accuracy'], color='skyblue')
plt.title('Comparison of Algorithms for Heart Disease Prediction')
plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set y-axis limit to ensure visibility of differences
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Accuracy scores for Heart Disease, Diabetes, and Cancer
heart_disease_accuracy = 0.8852459016393442
diabetes_accuracy = 0.9122807017543859
cancer_accuracy = 0.9766081871345029

# Create a DataFrame for the accuracies
accuracies_df = pd.DataFrame({
    'Disease': ['Heart Disease', 'Diabetes', 'Cancer'],
    'Accuracy': [heart_disease_accuracy, diabetes_accuracy, cancer_accuracy]
})

# Plot the comparison
plt.figure(figsize=(8, 6))
plt.bar(accuracies_df['Disease'], accuracies_df['Accuracy'], color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Comparison of Best Models for Different Diseases')
plt.xlabel('Disease')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set y-axis limit to ensure visibility of differences
plt.show()
```
